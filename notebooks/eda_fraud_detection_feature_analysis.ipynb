```python
# Import necessary modules
import sys
sys.path.append('../fraud_detection')  # Adjust this path based on your working directory
from utils.logging_util import setup_logger



from feature_engineering import preprocess_transaction_data,standardize_numerical_features,label_encode_categorical_columns

from utils.data_loader import load_datasets
from data_preprocessing import preprocess_data,preprocess_ip_data
from utils.visualization_util import plot_transaction_amount_distribution, plot_fraudulent_vs_non_fraudulent, plot_purchase_value_distribution, plot_age_distribution, plot_bivariate_analysis


# File paths
creditcard_file_path = '../data/creditcard.csv'
fraud_data_file_path = '../data/Fraud_Data.csv'
ip_to_country_file_path = '../data/IpAddress_to_Country.csv'

# Setup logger to write logs to 'eda_fraud_detection.log'
logger = setup_logger('../logs/eda_fraud_detection.log')
```

```python


creditcard_df, fraud_data_df, ip_to_country_df = load_datasets(
    creditcard_file_path, fraud_data_file_path, ip_to_country_file_path
)

logger.info("Data loading completed.")

# Log data preprocessing
logger.info("Starting preprocessing data...")
creditcard_df, fraud_data_df, ip_to_country_df = preprocess_data(
    creditcard_df, fraud_data_df, ip_to_country_df
)

print("Credit Card Dataset Info:")
print(creditcard_df.info())
print("Fraud Data Dataset Info:")
print(fraud_data_df.info())
print("IP to Country Dataset Info:")
print(ip_to_country_df.info())
```
